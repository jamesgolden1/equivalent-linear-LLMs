```python
#@title Run Jacobian Analysis
%%time
# %%writefile run_llama.py
import os
import gc
import torch
import argparse

# setattr(JacobianAnalyzer, 'model_forward', model_forward)

# Initialize the analyzer
# analyzer = JacobianAnalyzer(model_name="Qwen/Qwen3-8B")
analyzer = JacobianAnalyzer(model_name="deepseek-ai/DeepSeek-R1-0528-Qwen3-8B")
run_all = False

# text = 'The Golden'
# text = 'To build, they need'
text = 'The bridge out of Marin is the'
max_new_tokens=1
temperature=1e-6

# Generate output
analyzer.generate(text, max_new_tokens, temperature);

# # # Compute Jacobian
analyzer.compute_jacobian()
analyzer.compute_jacobian_nonlinear()
analyzer.plot_jacobian_comparison(text,filename_png="fig3")

analyzer.compute_jacobian_row_col_norm(n_components=8)#, svs=1)
analyzer.plot_singular_values(mode="row_col_vectors",filename_png="fig4_col")

analyzer.compute_jacobian_svd(n_components=8, svs=1, token_list=list(range(analyzer.embeds.shape[1])))
analyzer.plot_singular_values(filename_png="fig4_svd")
```

```python
%%time
# layers
layerlist = list(range(2,len(analyzer.model.model.layers)))

key = 'layer'
# layerlist = [len(analyzer.model.model.layers)-2]
analyzer.compute_jacobian_layers_svd(token_list=list(range(analyzer.embeds.shape[1])),fast_approx=False,layer_mode='cumulative',tokens_combined=False,layerlist=layerlist,n_components=3,svs=3,key=key)#,transform_to_output=False,transform_to_last_layer=len(analyzer.model.model.layers)-0,key=key)#,filename="fig5_svd_layers_llama_3_2")
```

```python
df=[]
ctr=0
svs = 3
for ii in range(34):
  df.append([])
  for jj in range(7):
    df[-1].append([])
    for sv in range(svs):
      df[-1][-1].append(analyzer.usvec_layers['layer'][ctr][sv])
    ctr+=1
```

```python
df
```

```python
ntks=7
dfrs = np.array(df).swapaxes(1, 2).reshape([34*svs,ntks])
df=pd.DataFrame([
    [''.join(c if ord(c) < 128 else '' for c in text).replace(",","").replace("-","").replace("_","").replace("*","") for text in li[:]]
    for li in dfrs
],index=["Layer "+str(ii//svs)+'_'+str(ii%svs) for ii in range(34*svs)],columns=["Input token "+str(ii) for ii in range(ntks)])#.to_csv("layers_cumulative_no_proj_top1.csv")
    # ],index=[str(ii//svs)+'_'+str(ii%svs) for ii in range(30*svs)])#.to_csv("layers_cumulative_no_proj_top4.csv")
```
